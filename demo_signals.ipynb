# Visualizing audio signals
### Import the audio and plot it

```python
fs,x = read('audiofile.wav')
x = x[:,0] # Convert to mono
N = x.shape[0]

# Plot audio 
plt.figure(figsize=(10,4))
plt.plot(x)
plt.title('Plot of audio amplitude over time')
plt.xlabel('samples')
plt.show()
```

### Zoom in on a window of the signal

In the following code block, you can manipulate the start index and the length of the window to zoom in on different parts of the signal.

```python
start_index = 50000 # Change this!
window_length = 20 # Change this!

plt.figure(figsize=(10,4))
plt.stem(x[start_index:start_index+window_length])
plt.title('Plot of audio amplitude over time, zoomed in')
plt.xlabel('samples')
plt.show()
```

# Unit impulse and shifted unit impulse

$ x_0 = \delta[n] $

$ x_1 = \delta[n-m] $

The unit impulse function $\delta[n]$ has a value of $1$ at $n = 0$ and a value of $0$ everywhere else. We can shift the unit impulse by introducing a new value, $m$.

The following code block plots the unit impulse within the range of $-5 \leq n \leq 5$. Thus you can choose values of $m$ between $-5$ and $5$ to see how the impulse is shifted.

```python
n = np.linspace(-5,5,11)
m = 2 # Change this!

x0 = np.zeros_like(n)
x1 = np.zeros_like(n)
x0[5] = 1
x1[5+m] = 1
plt.figure(figsize=(8,8))
plt.subplot(211)
plt.stem(n,x0)
plt.title('x1')
plt.subplot(212)
plt.stem(n,x1)
plt.title('x2')
plt.show()
```

# Unit impulse using Scipy package

$ x_2 = \delta[n] $

$ x_3 = \delta[n-m] $

The Scipy function will generate a unit impulse, and has an optional parameter to introduce a shift. You can learn more by looking at Scipy documentation. Here, you can once again choose $m$, this time a positive integer up to $10$.

```python
from scipy.signal import unit_impulse
n = np.linspace(0,10,11)
m = 2 # Change this!
x2 = unit_impulse(11)
x3 = unit_impulse(11,m)
plt.figure(figsize=(8,8))
plt.subplot(211)
plt.stem(n,x2)
plt.title('x2')
plt.subplot(212)
plt.stem(n,x3)
plt.title('x3')
plt.show()
```

# Sinusoids in Python
A simple sinusoid with frequency $f$.

$ x_4 = \sin(2 \pi f n) $

This code block plots a sinusoid with a window length of 2. You can change the value of $f$ to modulate the frequency of the sinusoid.

```python
n = np.linspace(0,2,1001)
f = 1 # Change this!
x4 = np.sin(2*np.pi*f*n)
plt.plot(n,x4)
plt.xticks([0,1,2])
plt.show()
```

# Sounding out middle C
Middle C (or C4) has a frequency of 261.6 Hz.

$ x = \sin(2\pi\cdot261.6\cdot n)$

Here we set a variable $f_1$ to the desired frequency. You can change the value to sound out different notes. For example, Middle A is 440 Hz.

```python
fs = 4000
n = np.linspace(0,2,2*fs,endpoint=False)
f1 = 261.6 # Change this number for different notes
x = np.sin(2*np.pi*f1*n)
plt.plot(n[:100],x[:100])
plt.xlabel('seconds')
plt.show()
```

```python
Audio(x, rate = fs)
```
